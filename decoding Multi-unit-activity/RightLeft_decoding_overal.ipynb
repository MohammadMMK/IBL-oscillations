```python
import json
with open('pid_eid_pairs.json', 'r') as f:
    pid_eid_pairs = json.load(f)
print(f"Number of sessions with passive data: {len(pid_eid_pairs)}")
```

```python
import os 
import numpy as np
import pandas as pd
no_path = []
no_decoded = []
acronyms_all = []
depth_all = []
ids_all = []
decodability_scores_mean_all = []
decodability_scores_max_all = []
decodability_scores_min_all = []
pid_all = []
for pid, eid in pid_eid_pairs:
    base_path = '/mnt/data/AdaptiveControl/IBLrawdata/classification/preprocess_firingRate'
    path = os.path.join(base_path, f'{pid}.pkl')
    if not os.path.exists(path):
        no_path.append(pid)
        continue
    import pickle
    with open(path, 'rb') as f:
        data = pickle.load(f)
    decoded = [ key for key in data.keys() if 'decodability_scores' in key]
    if len(decoded) == 0:
        no_decoded.append(pid)
        continue
    depth_info = data['depth_info']
    acronyms = depth_info['acronyms']
    depth = depth_info['depth']
    ids = depth_info['ids']
    decodability_scores = data['decodability_scores']
    decodability_scores_mean = decodability_scores.mean(axis=1)
    decodability_scores_max = decodability_scores.max(axis=1)
    decodability_scores_min = decodability_scores.min(axis=1)
    pids = [pid] * len(acronyms)


    decodability_scores_mean_all.extend(decodability_scores_mean)
    acronyms_all.extend(acronyms)
    decodability_scores_max_all.extend(decodability_scores_max)
    pid_all.extend(pids)
    decodability_scores_min_all.extend(decodability_scores_min)
    depth_all.extend(depth)
    ids_all.extend(ids)

data = {'pid': pid_all,
    'acronyms': acronyms_all,
    'depth': depth_all,
    'id': ids_all,
    'decodability_scores_mean': decodability_scores_mean_all,
    'decodability_scores_max': decodability_scores_max_all,
    'decodability_scores_min': decodability_scores_min_all,
}
df = pd.DataFrame(data)
# save df 
df.to_csv('decodability_scores.csv', index=False)    
```

```python
import pandas as pd

# Load df
df = pd.read_csv('decodability_scores.csv')

# Filter the data where decodability_scores_max > 0.6
filtered_df = df[df['decodability_scores_max'] > 0.6]

# Define visual regions
visual_regions = ['VISp', 'VISpm', 'VISam', 'VISa', 'VISrl', 'VISal', 'VISli', 'VISl']
# Create a regex pattern to match any of the visual regions
pattern = '|'.join(visual_regions)

# Filter the data to only include rows where the acronym contains any visual region
filtered_df = filtered_df[filtered_df['acronyms'].str.contains(pattern)]
# Count occurrences of each acronym in the filtered data
summary_table = filtered_df['acronyms'].value_counts().reset_index()
summary_table.columns = ['acronyms', 'count_over_0.6']

# Display the summary table
display(summary_table)
```
