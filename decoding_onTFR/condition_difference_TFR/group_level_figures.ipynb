```python
import os
from PIL import Image
import matplotlib.pyplot as plt

# Define the path
path = '/crnldata/cophy/TeamProjects/mohammad/ibl-oscillations/_analyses/_IBLworkflows/decoding/figures'

# Get the list of raw and csd files
raw_files = [f for f in os.listdir(path) if 'raw' in f]
csd_files = [f for f in os.listdir(path) if 'csd' in f]

# Function to display images
def display_images(file_list):
    file_list.sort()
    plt.figure(figsize=(20, 25))
    
    for i, file in enumerate(file_list):
        img_path = os.path.join(path, file)
        img = Image.open(img_path)
        plt.subplot(1, len(file_list), i + 1)
        plt.imshow(img)
        plt.axis('off')
    plt.show()
# Display raw images
display_images(raw_files)

# # Display csd images
display_images(csd_files)
```
